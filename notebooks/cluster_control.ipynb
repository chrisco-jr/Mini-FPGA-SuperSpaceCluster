Import the cluster client library:

```python
import sys
sys.path.append('../platformio_slip/python_client')

from broccoli_cluster import BroccoliCluster
```

## Connect to Cluster

Connect to the ESP32 master node via serial port:

```python
# Change COM23 to your actual port
cluster = BroccoliCluster('COM23')
cluster.connect()
```

## Define Tasks

Define computational tasks that will run on the ESP32 workers.
Tasks are defined as simple expressions using argument names.

```python
# Define an addition task
cluster.define_task('add', 'x + y')

# Define a multiplication task
cluster.define_task('mul', 'x * y')

# Define a power task
cluster.define_task('power', 'x * x')
```

List all defined tasks:

```python
tasks = cluster.list_tasks()
print("Available tasks:")
for task in tasks:
    print(f"  - {task}")
```

## Execute Tasks

Submit tasks to the cluster and get results:

```python
# Execute addition task
result = cluster.execute('add', 5, 3)
print(f"5 + 3 = {result}")
```

```python
# Execute multiplication task
result = cluster.execute('mul', 7, 6)
print(f"7 * 6 = {result}")
```

```python
# Execute power task
result = cluster.execute('power', 12)
print(f"12^2 = {result}")
```

## Batch Execution

Execute multiple tasks in sequence:

```python
# Compute multiple values
print("Computing batch of tasks...\n")

for i in range(1, 6):
    result = cluster.execute('power', i)
    print(f"{i}^2 = {result}")
```

## Advanced: Define Complex Tasks

You can define more complex tasks:

```python
# Define a task that computes average (note: worker needs to support division)
cluster.define_task('avg', '(x + y) / 2')

# Test it
result = cluster.execute('avg', 10, 20)
print(f"Average of 10 and 20 = {result}")
```

## Statistics

Check SLIP communication statistics:

```python
cluster.stats()
```

## Disconnect

When done, disconnect from the cluster:

```python
cluster.disconnect()
```

## Example: Using Context Manager

You can also use the cluster with a context manager for automatic connect/disconnect:

```python
with BroccoliCluster('COM23') as cluster:
    # Define tasks
    cluster.define_task('add', 'x + y')
    
    # Execute
    result = cluster.execute('add', 100, 200)
    print(f"Result: {result}")
    
# Automatically disconnected here
```
